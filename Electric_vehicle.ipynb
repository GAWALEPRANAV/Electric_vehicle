```python
plt.subplots(figsize=(12,6))
a=sns.countplot(data=EV,x='Model Year',order=EV['Model Year'].value_counts().index)
plt.show()
```

```python
Car_Popularity = EV.value_counts(["Make","Model"],ascending=True).tail(10)
Car_Popularity
```

```python
Car_Popularity.plot(kind='barh',title = 'Car Model Popularity Graph',figsize=(20,10),
                             ylabel = "TOP 10 POPULAR MODELS", xlabel = "Total number of cars", fontsize = 15,color='lime')
plt.show()
```

```python
CarsPerCounty = EV.value_counts(["County"],ascending=False).head(10)
CarsPerCounty
```

```python
CarsPerCounty.plot(kind='barh', title = 'Cars Per County Graph',figsize=(15,10),fontsize = 15,
                            ylabel = "County", xlabel = "Total number of cars",color='magenta')
plt.show()
```

```python
CarsPerCityandCounty=EV.value_counts(['County','City'],ascending=False)
CarsPerCityandCounty
```

```python
CarsPerCityandCounty.head(15).plot(kind='barh',xlabel='Total number of Cars',ylabel='County and City',
                                   title='Cars Per City and County Graph',figsize=(15,10),fontsize=15,color='teal')
plt.show()
```

```python
EV['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts().plot.pie(ylabel='Clean Alternative Fuel Vehicle (CAFV) Eligibility',autopct='%1.1f%%',colors=['c','m','b'],figsize=(15,10),fontsize=15)
plt.show()
```
